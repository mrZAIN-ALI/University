```python
# Define the content of the dataset.yaml file
yaml_content = """
train: /content/dataset/images/train
val: /content/dataset/images/val

nc: 22

names:
  - CompanyName
  - SerialNumber
  - Date
  - SupplierName
  - BuyerName
  - SupplierAddress
  - BuyerAddress
  - SupplierRegNo
  - BuyerRegNo
  - SupplierNTN
  - BuyerNTN
  - TermOfSale
  - Products
  - Quantity
  - Rate
  - AmountExcludingTax
  - SalesTax
  - AmountIncludingTax
  - TotalSalesTax
  - TotalAmount
  - Signature
  - NameDesignation
"""
```

```python
# Define the file path where you want to save the YAML file
yaml_file_path = '/content/yolov5/dataset.yaml'

# Write the YAML content to the file
with open(yaml_file_path, 'w') as file:
    file.write(yaml_content)

print(f"YAML file created at {yaml_file_path}")
```

```python
import os

# Path to the annotations folder
annotations_path = "/content/dataset/labels"

# Check and create missing folders
for split in ["train", "val"]:
    split_folder = os.path.join(annotations_path, split)
    os.makedirs(split_folder, exist_ok=True)  # Ensure folders exist

# Iterate through all .txt files
for split in ["train", "val"]:
    split_folder = os.path.join(annotations_path, split)
    for file_name in os.listdir(split_folder):
        if file_name.endswith(".txt"):
            file_path = os.path.join(split_folder, file_name)

            # Read and adjust annotations
            with open(file_path, "r") as f:
                lines = f.readlines()
            adjusted_lines = []
            for line in lines:
                parts = line.strip().split()
                parts[0] = str(int(parts[0]) - 1)  # Subtract 1 from class_id
                adjusted_lines.append(" ".join(parts))

            # Write back the adjusted annotations
            with open(file_path, "w") as f:
                f.write("\n".join(adjusted_lines))

print("Annotations adjusted successfully!")
```
